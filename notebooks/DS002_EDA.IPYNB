```python
import os
import pandas as pd
import matplotlib.pyplot as plt

DATA_PATH2 = os.getenv("DATA_PATH2")
DATA_PATH4 = os.getenv("DATA_PATH4")
```

```python
df = pd.read_csv(DATA_PATH2, delimiter=',', encoding='utf-8')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

### Null Manage

```python
df.fillna(value={'artists': 'Unknown Artist', 'album_name': 'Unknown Album', 'track_name': 'Unknown Track'}, inplace=True)

df.info()
```

```python
for i in df.columns:
    if df[i].dtype == 'object':
        df[i] = df[i].astype('string')

df.info()
```

```python
to_eliminate = ['Unnamed: 0', 'track_id']
df = df.drop(columns= to_eliminate)

df.info()
```

```python
df.head()
```

```python
df.to_csv(DATA_PATH4, index= False)
```

```python
artist_count = df['artists'].value_counts()
top_artists = artist_count.head(10)

plt.pie(top_artists, labels= top_artists.index, autopct= '%1.1f%%', startangle= 140)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', prop={'size': 6})
plt.title('Artist')
plt.show()
```
